```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
from sklearn.linear_model import LinearRegression
import seaborn as sns
import statsmodels.api as sm

# Lista de nombres de archivos
archivos = ["grupo1.csv", "grupo2.csv", "grupo3.csv"]

# Bucle para el proceso  de cada conjunto de datos
for i, archivo in enumerate(archivos):
    print(f"==================================================")
    print(f"|              Análisis de {archivo}             |")
    print(f"==================================================")

    # Carga de Datos
    df = pd.read_csv(archivo)
    df.columns = ['HorasEstudio', 'Calificacion']
    X = df[['HorasEstudio']]
    y = df['Calificacion']

    print("\n### 1. Estadísticas ###")
    print(df.describe().round(2))

    # Genero Gráficos

    ## Histograma de Calificaciones
    plt.figure(figsize=(8, 5))
    sns.histplot(df['Calificacion'], kde=True)
    plt.title(f'Histograma de Calificaciones para {archivo}')
    plt.xlabel('Calificación Final')
    plt.ylabel('Frecuencia')
    plt.grid(axis='y', alpha=0.5)
    plt.show() # Muestra el primer gráfico

    # Regresión Lineal

    model = LinearRegression()
    model.fit(X, y)
    pendiente = model.coef_[0]
    intercepto = model.intercept_

    # Usar statsmodels para el R-cuadrado y valores P
    X_sm = sm.add_constant(X)
    model_sm = sm.OLS(y, X_sm).fit()
    coef_correlacion, p_value_pearson = pearsonr(df['HorasEstudio'], df['Calificacion'])

    print("\n### 2. Resultados del Modelo y Correlación ###")
    print(f"Ecuación: Calificación = {pendiente:.3f} * HorasEstudio + {intercepto:.3f}")
    print(f"Coeficiente de Correlación (r): {coef_correlacion:.4f}")
    print(f"R-cuadrado Ajustado: {model_sm.rsquared_adj:.4f}")
    print(model_sm.summary())

    # Gráfico de Regresión

    plt.figure(figsize=(10, 6))
    sns.scatterplot(x='HorasEstudio', y='Calificacion', data=df, label='Datos Originales')
    plt.plot(X, model.predict(X), color='red', label='Línea de Regresión', linewidth=2)
    plt.title(f'Regresión Lineal: Horas vs. Calificación ({archivo})')
    plt.xlabel('Horas de Estudio')
    plt.ylabel('Calificación Final')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.show() # Muestra el segundo gráfico



    #  Interpretación de Resultados

    print("\n### 3. Interpretación de Preguntas ###")

    r_squared = model_sm.rsquared_adj
    p_value_slope = model_sm.pvalues['HorasEstudio']
    es_significativo = p_value_slope < 0.05

    # 1. Coeficiente de Regresión (Pendiente)
    print(f"1. Interpretación del Coeficiente de Regresión (Pendiente = {pendiente:.3f}):")
    print(f"   Por cada hora adicional de estudio, se espera que la calificación {'aumente' if pendiente > 0 else 'disminuya'} en aproximadamente {abs(pendiente):.3f} puntos.")

    # 2. Coeficiente de Correlación (r)
    fuerza = "muy fuerte" if abs(coef_correlacion) >= 0.9 else ("fuerte" if abs(coef_correlacion) >= 0.7 else ("moderada" if abs(coef_correlacion) >= 0.5 else ("débil" if abs(coef_correlacion) >= 0.3 else "muy débil/inexistente")))
    direccion = "positiva" if coef_correlacion > 0.1 else ("negativa" if coef_correlacion < -0.1 else "nula")
    print(f"\n2. Interpretación del Coeficiente de Correlación (r = {coef_correlacion:.4f}):")
    print(f"   Indica una relación lineal **{fuerza}** y de dirección **{direccion}**. (Cercano a $\\pm 1$ es fuerte, cercano a $0$ es débil/nula).")

    # 3. Significancia Estadística
    significancia = "SÍ es estadísticamente significativo" if es_significativo else "NO es estadísticamente significativo"
    print(f"\n3. ¿El modelo es estadísticamente significativo? - {significancia}:")
    print(f"   El valor $p$ de la pendiente ({p_value_slope:.4f}) es {'menor' if es_significativo else 'mayor'} a $0.05$.")

    # 4. Confiabilidad para Predecir
    print(f"\n4. Confiabilidad para Predecir (R² Ajustado = {r_squared:.4f}):")
    print(f"   El modelo explica el **{r_squared*100:.2f}%** de la variación en las calificaciones. La confiabilidad es {'**alta**' if r_squared >= 0.7 else ('**moderada**' if r_squared >= 0.5 else '**baja**')}.")

    print("\n" + "="*70 + "\n")
```

Esta práctica de regresión lineal fue fundamental para entender que no basta con calcular una ecuación; lo más importante es interpretar los resultados para saber si el modelo sirve en la vida real.

Los tres grupos mostraron escenarios muy diferentes:

* la Correlación ($r$) es la que manda. Es lo primero que debo checar para saber si el modelo será útil.

Grupo 3: Tenía una correlación casi perfecta ($r=0.98$). Esto significa que las Horas de Estudio y la Calificación están conectadas de forma directa y muy fuerte. Este fue el caso ideal.

Grupo 1: Su correlación estaba prácticamente en cero ($r \approx -0.09$) y además no fue significativa. Aprendí que si la correlación es nula, la regresión lineal es un modelo inútil para predecir, porque la relación es puro azar.

* El Coeficiente de Determinación ($R^2$) me dice qué tan confiable es mi modelo para predecir.En el Grupo 3, el $R^2$ fue altísimo ($\approx 96\%$). Esto significa que el $96\%$ de las notas se explican SOLO por las horas de estudio. El modelo es un predictor excelente.

En el Grupo 2, la confiabilidad es buena ($\approx 70\%$). El estudio es muy importante, pero hay un $\approx 30\%$ de la calificación que depende de otras variables no medidas.

En el Grupo 1, el $R^2$ fue $0\%$. No importa cuántas horas estudiaron, el modelo no pudo predecir nada.

 El "retorno de inversión" de cada hora de estudio:Grupos 2 y 3: Al ser positivas, confirman que estudiar SÍ paga. Por ejemplo, en el Grupo 3, una hora extra te da $\approx 5.5$ puntos más.

 Grupo 1: La pendiente fue negativa y sin significancia, lo que confirma que el tiempo de estudio en este grupo no tiene un efecto real en la calificación.

 En resumen: Pude aplicar el modelo, pero lo más importante es que sé que tengo que mirar el valor $p$ para saber si la relación es real y el $R^2$ para saber si vale la pena usar el modelo para hacer predicciones.
