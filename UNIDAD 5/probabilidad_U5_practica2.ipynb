El coeficiente de determinaci√≥n es la principal forma en que podemos medir el grado, o fuerza, de la asociaci√≥n que existe entre dos variables, X y Y.

‚Ä¢  El coeficiente de correlaci√≥n es la segunda medida que podemos usar para describirqu√© tan bien explica una variable a otra.

‚Ä¢ Cuando tratamos con muestras, el coeficiente de correlaci√≥n de la muestra se denota por r y es la ra√≠z cuadrada del coeficiente de determinaci√≥n de muestra:

r significa que tanto (%) los datos se relacionan entre s√≠.

El signo de r indica la direcci√≥n de la relaci√≥n entre las dos variables X y Y.

‚Ä¢ Si existe una relaci√≥n inversa, es decir, si Y disminuye al aumentar X, entonces r caer√° entre 0 y - 1.

‚Ä¢ De manera similar, si existe una relaci√≥n directa (si Y aumenta al aumentar X), entonces r ser√° un valor en el intervalo de 0 a 1.


El coeficiente de correlaci√≥n de Pearson puede variar entre -1 y 1:
‚Ä¢ Un coeficiente de correlaci√≥n de 1 indica una relaci√≥n lineal
positiva perfecta.

‚Ä¢ Un coeficiente de correlaci√≥n de -1 indica una relaci√≥n lineal negativa perfecta.

‚Ä¢ Un coeficiente de correlaci√≥n de 0 indica que no hay relaci√≥n lineal entre las variables.

El valor p ayuda a determinar si lacorrelaci√≥n observada es estad√≠sticamente
significativa.

‚Ä¢ Un valor p peque√±o (t√≠picamente < 0.05) indica que la correlaci√≥n es significativa.

```python
from scipy.stats import pearsonr

# Datos de entrada
publicidad = [2, 3, 5, 4, 11, 5]
ganancias = [20, 25, 34, 30, 40, 31]

# Calcular el coeficiente de correlaci√≥n de Pearson
coeficiente_correlacion, p_valor = pearsonr(publicidad, ganancias)

print("Coeficiente de correlaci√≥n de Pearson:", coeficiente_correlacion)
print("Valor p:", p_valor)
```
