```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
AUTOS=pd.read_csv("autos.csv")
```

```python
precio_min = AUTOS["price"].min()
print(f"Valor mínimo de precio: {precio_min}")
```

```python
precio_max = AUTOS["price"].max()
print(f"Valor máximo de precio: {precio_max}")
```

```python
precio_moda = AUTOS["price"].mode()[0]
print(f"Moda de precios: {precio_moda}")
```

```python
precio_media = AUTOS["price"].mean()
print(f"Media de precios: {precio_media:.2f}")
```

```python
precio_mediana = AUTOS["price"].median()
print(f"Mediana de precios: {precio_mediana}")
```

```python
precio_std = AUTOS["price"].std()
print(f"Desviación estándar: {precio_std:.2f}")
```

```python
print("\nEstadísticos con describe():")
display(AUTOS["price"].describe())
```

```python
n = len(AUTOS["price"])
bins_sturges = int(1 + 3.322 * np.log10(n))
print(f"Número óptimo de bins (Sturges): {bins_sturges}")
```

```python
plt.figure(figsize=(10,6))
plt.hist(AUTOS["price"], bins=bins_sturges, edgecolor="black", alpha=0.7)
plt.title("Histograma de Precios de Automóviles", fontsize=14)
plt.xlabel("Precio")
plt.ylabel("Frecuencia")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x="body-style", y="price", data=AUTOS)
plt.title("Relación entre Precio y Tipo de Vehículo (Body Style)", fontsize=14)
plt.xlabel("Tipo de Vehículo")
plt.ylabel("Precio")
plt.xticks(rotation=30)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x="drive-wheels", y="price", data=AUTOS)
plt.title("Relación entre Precio y Tracción de Ruedas (Drive Wheels)", fontsize=14)
plt.xlabel("Tracción de Ruedas")
plt.ylabel("Precio")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x="engine-location", y="price", data=AUTOS)
plt.title("Relación entre Precio y Ubicación del Motor (Engine Location)", fontsize=14)
plt.xlabel("Ubicación del Motor")
plt.ylabel("Precio")
plt.show()
```

```python
print("\n=== Conclusiones ===")
print("- Los precios de los autos varían desde", precio_min, "hasta", precio_max)
print("- La media es mayor que la mediana, lo que indica que la distribución de precios está sesgada a la derecha (hay autos muy caros que elevan la media).")
print("- El histograma muestra la concentración de autos en rangos de precios bajos-medios.")
print("- Los boxplots revelan diferencias claras en el precio según el estilo de carrocería, tipo de tracción y ubicación del motor.")
```
