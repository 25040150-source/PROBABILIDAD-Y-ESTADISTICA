```python
import random
import matplotlib.pyplot as plt
from statistics import *
```

```python
fabrica1=[]
for i in range(100):
  numero_aleatorio=round(random.uniform(50,100),2)
  fabrica1.append(numero_aleatorio)
print(fabrica1)
```

```python
fabrica2=[]
for i in range(100):
  numero_aleatorio2=round(random.uniform(40,90),2)
  fabrica2.append(numero_aleatorio2)
print(fabrica2)
```

```python
plt.boxplot(fabrica1)
plt.ylabel('produccion en mts.')
plt.title("produccion de la fabrica1")
plt.show()
```

```python
plt.boxplot(fabrica2)
plt.ylabel('produccion en mts.')
plt.title("produccion de la fabrica2")
plt.show()
```

```python
fig,(ax1,ax2)=plt.subplots(nrows=1,ncols=2,figsize=(9,4))
bplot1=ax1.boxplot(fabrica1,
                   vert=True,
                   patch_artist=True)
ax1.set_title("PRODUCCION FABRICA 1")
bplot1=ax2.boxplot(fabrica2,
                   vert=True,
                   patch_artist=True)
ax1.set_title("PRODUCCION FABRICA 2")
plt.show()
```
