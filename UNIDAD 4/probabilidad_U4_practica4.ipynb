El número promedio de accidentes de tránsito en cierto crucero de carretera es dos por semana. Suponga
que el número de accidentes sigue una distribución de Poisson con una media = 2.
1. Encuentre la probabilidad de que no haya accidentes en este crucero de carretera durante un periodo
de 1 semana.
2. Encuentre la probabilidad de que a lo sumo haya tres accidentes en esta sección de carretera durante un
periodo de 2 semanas.

```python
from scipy.stats import poisson

#parametros de la distribución de Poisson
media_accidentes = 2

#parte 1: probabilidad de que no haya accidentes en una semana
probabilidad_0_accidentes = poisson.pmf( 0, media_accidentes)
print("1 Probabilidad de que no haya accidentes en una semana: ", probabilidad_0_accidentes)

#parte 2 probabilidad de que a lo sumo haya 3 accidentes en 2 semanas
#convertimos la media a una media para dos semanas
media_2_semanas = media_accidentes * 2
#calculamos la probabilidad acumulada de hasta 3 accidentes en dos semanas
probabilidad_3_o_menos_accidentes = poisson.cdf(3,media_2_semanas)
print("2 Probabilidad de que a lo sumo haya 3 accidentes en 2 semanas: ", probabilidad_3_o_menos_accidentes)
```

Una compañía de seguros de vida asegura las vidas de 5,000 hombres de 42 años de edad.
Si estudios actuariales muestran que la probabilidad de que cualquier hombre de 42 años muera en un año determinado es .001, encuentre la probabilidad exacta de que la compañía tendrá que pagar x=4 reclamaciones durante un año determinado

```python
from scipy.stats import poisson

#parametros de la distribución de Poisson
Media_muertes = 5000 * .001

#numero de reclamaciones
x= 4

#calcular la probabilidad de que la compañía tenga que pagar 4 reclamaciones en un año determinado
probabilidad_4_reclamaciones = poisson.pmf(x,Media_muertes)

print("la probabilidad de que la compañía tenga que pagar 4 reclamaciones es de: ", probabilidad_4_reclamaciones)
```
