ejercicio 3:
*para una variable aleatoria binomial n=10 y una probabilidad de 0.10.

*calcular las probabilidades para cada valor de x= 0 hasta 10.

*imprimir la tabla de valores x y sus probabilidades.

*graficar el histograma de la distribucion binomial.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom
#parametros de la distribucion binomial
n= 10
p= 0.10
#calcular las probabilidades para cada valor de x de 0 a 10
valores_x= np.arange(0,n+1)
probabilidades_x= [binom.pmf(x,n,p) for x in valores_x]

#imprimir la tabla de valores de x y sus probabilidades
print("valores de x\t probabilidades")
for x,prob in zip(valores_x,probabilidades_x):
    print(f"{x}\t\t{prob}")
```

```python
#graficar el histograma de la distribucion binomial
plt.figure(figsize=(8,6))
plt.bar(valores_x,probabilidades_x,color="skyblue")
plt.xlabel("valores de x")
plt.ylabel("probabilidades")
plt.title("distribucion binomial")
plt.xticks(valores_x)
plt.grid(True)
plt.show()
```
