problema 1: Una revisión de los pesos de 500 estudiantes del ITD, evidenció que la media es de 70 kilogramos y la desviación estándar es de 4 kilogramos. Suponiendo que los valores se distribuyen normalmente. Hallar lo siguiente:

*La probabilidad de estudiantes que pesan más de 68 y menos de 72 kg.

*Cuantos estudiantes pesan más de 68 y menos de 72 kg.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
# parámetros de la distribución normal
media = 70  # media
desv_std = 4  # desviación estándar
total_estudiantes = 500

# límites de peso
peso_inferior = 68
peso_superior = 72

# calculamos la probabilidad de que un estudiante pese menos de 72 kg
prob_menos72kg = stats.norm.cdf(peso_superior, media, desv_std)

# calculamos la probabilidad de que un estudiante pese menos de 68 kg
prob_menos68kg = stats.norm.cdf(peso_inferior, media, desv_std)

# la probabilidad de que un estudiante pese entre 68 y 72 kg es la diferencia
prob_entre68y72kg = prob_menos72kg - prob_menos68kg

# cálculo del número de estudiantes que pesan entre 68 y 72 kg
estudiantes_entre68y72kg = prob_entre68y72kg * total_estudiantes

print(f"Probabilidad de que un estudiante pese entre {peso_inferior}kg y {peso_superior}kg: {prob_entre68y72kg:.4f}")
print(f"Número de estudiantes que pesan entre {peso_inferior}kg y {peso_superior}kg: {estudiantes_entre68y72kg:.2f}")

# graficar la distribución normal
x = np.linspace(media - 4 * desv_std, media + 4 * desv_std, 1000)
y = stats.norm.pdf(x, media, desv_std)

plt.figure(figsize=(8, 5))
plt.plot(x, y, 'b', linewidth=2)
# sombrear el área entre 68kg y 72kg
plt.fill_between(x, y, where=((x >= peso_inferior) & (x <= peso_superior)), color='skyblue', alpha=0.5)
plt.title(f'Probabilidad de que un estudiante pese entre {peso_inferior}kg y {peso_superior}kg \n $\\mu={media}$, $\\sigma={desv_std}$, n={total_estudiantes}')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```
