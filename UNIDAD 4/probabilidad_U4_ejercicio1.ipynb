ejercicio 1: una revision de los pesos de 500 estudiantes del itd, evidencio que la media es de 70 kg y la desviacion std es de 4 kg. suponiendo que los valores se distribuyen normalmente;hallar lo siguiente:

*generar aleatoriamente las edades de los 500 estudiantes con una meida de 70kg, y una media de 4kg  
      
*realiza un histograma de las edades

*graficar la funcion de densidadde probabilidad(PDF).


```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
#generar datos para la distribucion normal
media = 70 #media
desv_std = 4 #desviacion estandar
num_datos = 500

#generamos los datos de manera aleatoria y los ponemos en una lista (array)
datos = np.random.normal(media,desv_std,num_datos)
datos [0:50] #mostrar los primeros 50
```

```python
#crear un rango de valores para crear densidad
x = np.linspace(media - 3*desv_std,media + 3*desv_std,num_datos)
#calcular el pdf de la distribucion normal
pdf = 1/(desv_std * np.sqrt(2 * np.pi)) * np.exp( - (x - media)**2 / (2 * desv_std**2))
#para no imprimir toda la lista de valores (500),desplegamos los primeros 50
print("imprimir los primeros 50 valores de X ",x[0:50],"y los primeros 50 pdf",pdf[0:50])
```

```python
#graficar histograma
plt.figure (figsize=(6,4))
plt.hist (datos,bins=15,alpha=0.4,color='b')
plt.title('distribucion normal')
plt.xlabel('peso en kg de los estudiantes')
plt.grid(True)
plt.show()
```

```python
#visualizar la pdf
plt.figure (figsize=(6,4))
plt.plot(x,pdf,label='normal(media:{};ds{})'.format(np.round(media,2),np.round(desv_std,2)))
plt.title('funcion de densidad de probabilidad(PDF)- distribucion normal ')
plt.xlabel('pesos en kg de los estudiantes')
plt.ylabel('densidad de probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#graficar histograma
plt.figure (figsize=(6,4))
plt.hist (datos,bins=15,alpha=0.4,density=True,color='b')
plt.plot(x,pdf,'k',linewidth=2)
plt.title('distribucion normal')
plt.xlabel('peso en kg de los estudiantes')
plt.grid(True)
plt.show()
```

ejercicio 2: una revision de los pesos de 500 estudiantes del itd, evidencio que la media es de 70 kg y la desviacion std es de 4 kg. suponiendo que los valores se distribuyen normalmente;hallar lo siguiente:

*Cuantos estudiantes psean menos de 75kg

*Cuantos estudiantes psean menos de 62kg

*Cuantos estudiantes psean mas de 79kg.

primero necesitamos calcular la probabilidad de que un estudiante pese menos de 75kg utilizando la funcion de distribucion acumulativa(CDF) de la distribucion normal.

```python
#parametros de la distribucion normal
media = 70 #media
desv_std = 4 #desviacion estandar
total_estudiantes = 500
#variable para el calculo de la probabilidad de que un estudiante pesa menos de 75kg
peso_limite= 75
#calculamos la probabilidad de que un estudiante pese menos
#de 75 kg utilizando la funcion de distribucion acumulativa (CDF)  de la distribucion normal
from scipy.stats import norm
prob_menos75kg= norm.cdf(peso_limite,media,desv_std)
#calculo del numero de estudiantes que pesan menos de 76kg
estudiantes_menos75kg = prob_menos75kg * total_estudiantes
print("probabilidad de que un estudiante pese menos de75kg",prob_menos75kg)
print("numero de estudiantes que pesan menos de 75kg",estudiantes_menos75kg)

#graficar la distribucion normal
x= np.linspace(media - 4*desv_std,media + 4*desv_std,1000)
y= norm.pdf(x,media,desv_std)

plt.plot(x, y, 'b', linewidth=2)
plt.fill_between(x, y, where=(x <= peso_limite), color='skyblue', alpha=0.5)
plt.title('Probabilidad de que un estudiante pese menos de 75kg \n $\\mu=70$, $\\sigma=4$, x=500 ')
plt.xlabel('peso (kg)')
plt.ylabel('densidad de probabilidad')
plt.grid(True)
plt.show()
```

problema 1: Una revisión de los pesos de 500 estudiantes del ITD, evidenció que la media es de 70 kilogramos y la desviación estándar es de 4 kilogramos. Suponiendo que los valores se distribuyen normalmente. Hallar lo siguiente:

*La probabilidad de estudiantes que pesan más de 68 y menos de 72 kg.

*Cuantos estudiantes pesan más de 68 y menos de 72 kg.


```python
# parámetros de la distribución normal
media = 70  # media
desv_std = 4  # desviación estándar
total_estudiantes = 500

# límites de peso
peso_inferior = 68
peso_superior = 72

# calculamos la probabilidad de que un estudiante pese menos de 72 kg
prob_menos72kg = norm.cdf(peso_superior, media, desv_std)

# calculamos la probabilidad de que un estudiante pese menos de 68 kg
prob_menos68kg = norm.cdf(peso_inferior, media, desv_std)

# la probabilidad de que un estudiante pese entre 68 y 72 kg es la diferencia
prob_entre68y72kg = prob_menos72kg - prob_menos68kg

# cálculo del número de estudiantes que pesan entre 68 y 72 kg
estudiantes_entre68y72kg = prob_entre68y72kg * total_estudiantes

print(f"Probabilidad de que un estudiante pese entre {peso_inferior}kg y {peso_superior}kg: {prob_entre68y72kg:.4f}")
print(f"Número de estudiantes que pesan entre {peso_inferior}kg y {peso_superior}kg: {estudiantes_entre68y72kg:.2f}")

# graficar la distribución normal
x = np.linspace(media - 4 * desv_std, media + 4 * desv_std, 1000)
y = norm.pdf(x, media, desv_std)

plt.figure(figsize=(8, 5))
plt.plot(x, y, 'b', linewidth=2)
# sombrear el área entre 68kg y 72kg
plt.fill_between(x, y, where=((x >= peso_inferior) & (x <= peso_superior)), color='skyblue', alpha=0.5)
plt.title(f'Probabilidad de que un estudiante pese entre {peso_inferior}kg y {peso_superior}kg \n $\\mu={media}$, $\\sigma={desv_std}$, n={total_estudiantes}')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```
