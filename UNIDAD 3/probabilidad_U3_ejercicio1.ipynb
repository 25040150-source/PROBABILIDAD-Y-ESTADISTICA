Un investigador biomedico realiza un conteo del numero de celulas anormales observadas en una muestra de tejido tomada de distintos pacientes.
la variable aleatoria X representa el numero de celulas anormales encontradas en una muestra seleccionada al azar de una lamina. a partir de observaciones previas, se estima la siguiente distribucion de probabilidad:

#num de celulas anormales (X)  [0,1,2,3,4,5]
#probabilidad P(X) [0.10,0.25,0.30,0.20,0.10,0.05]

1.-verifique que las probabilidades sumen 1.

2.-elabore la distribucion de probabilidad X.

*el valor esperado E(X) (promedio de celulas anormales por muestra).  

*la varianza y la desviacion estandar.

4.-construya la funcion de distribucion acumulada F(X) y representala graficamente.

5.-interprete los resultados en el contexto biomedico:

E(X), es 2.10. Esto significa que en promedio se espera encontrar alrededor de 2 c√©lulas anormales por muestra. En t√©rminos biom√©dicos, representa el n√∫mero t√≠pico de c√©lulas anormales que se podr√≠an observar en una l√°mina seleccionada al azar.

Esto significa que los valores bajos son mucho m√°s probables, mientras que los valores altos (4 o 5) son poco frecuentes. En conclusi√≥n, la mayor√≠a de las muestras presentan pocos casos de c√©lulas anormales, lo que sugiere un comportamiento estable y poco riesgoso en las observaciones.

```python
# =========================================================
# EJERCICIO 1 - C√©lulas anormales por muestra
# =========================================================

import matplotlib.pyplot as plt
import numpy as np

# Valores de la variable aleatoria
x = np.array([0, 1, 2, 3, 4, 5])
p = np.array([0.10, 0.25, 0.30, 0.20, 0.10, 0.05])

# 1. Verificar suma de probabilidades
print("Suma de probabilidades:", np.sum(p))

# 2. Calcular valor esperado
E_X = np.sum(x * p)
print("Valor esperado E(X):", E_X)

# 3. Calcular varianza y desviaci√≥n est√°ndar
E_X2 = np.sum(x**2 * p)
Var_X = E_X2 - E_X**2
Desv_X = np.sqrt(Var_X)
print("Varianza:", Var_X)
print("Desviaci√≥n est√°ndar:", Desv_X)

# 4. Funci√≥n de distribuci√≥n acumulada
F = np.cumsum(p)
print("Funci√≥n acumulada F(x):", F)

# 5. Graficar FDA
plt.step(x, F, where='post', color='blue', label='F(x)')
plt.title("Funci√≥n de Distribuci√≥n Acumulada - Ejercicio 1")
plt.xlabel("N√∫mero de c√©lulas anormales (X)")
plt.ylabel("Probabilidad acumulada")
plt.grid(True)
plt.legend()
plt.show()
```
