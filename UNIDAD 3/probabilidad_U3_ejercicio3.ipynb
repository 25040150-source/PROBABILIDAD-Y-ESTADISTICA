¿cuantas unidades se espera vender por semana?

El valor esperado
𝐸(𝑋)
E(X) es 2.8 unidades, lo que significa que en promedio se espera vender aproximadamente 3 productos por semana.
Este valor representa la cantidad típica de ventas semanales observadas según la distribución de probabilidad obtenida.

¿la demanda es estable o muy variable?

La desviación estándar calculada fue 1.29, lo que indica que la demanda es relativamente estable.
Esto significa que las ventas suelen mantenerse cercanas al promedio (entre 2 y 4 unidades por semana), con poca dispersión entre los resultados. En conclusión, la tienda presenta un comportamiento de ventas constante, sin variaciones extremas en la demanda semanal.



```python
# =========================================================
# EJERCICIO 3 - Ventas semanales de un producto
# =========================================================

import matplotlib.pyplot as plt
import numpy as np

# Valores de la variable aleatoria
x = np.array([0, 1, 2, 3, 4, 5])
p = np.array([0.05, 0.10, 0.25, 0.30, 0.20, 0.10])

# 1. Verificar suma
print("Suma de probabilidades:", np.sum(p))

# 2. Valor esperado
E_X = np.sum(x * p)
print("Valor esperado E(X):", E_X)

# 3. Varianza y desviación estándar
E_X2 = np.sum(x**2 * p)
Var_X = E_X2 - E_X**2
Desv_X = np.sqrt(Var_X)
print("Varianza:", Var_X)
print("Desviación estándar:", Desv_X)

# 4. Función acumulada
F = np.cumsum(p)
print("Función acumulada F(x):", F)

# 5. Gráfica FDA
plt.step(x, F, where='post', color='red', label='F(x)')
plt.title("Función de Distribución Acumulada - Ejercicio 3")
plt.xlabel("Unidades vendidas (X)")
plt.ylabel("Probabilidad acumulada")
plt.grid(True)
plt.legend()
plt.show()
```
