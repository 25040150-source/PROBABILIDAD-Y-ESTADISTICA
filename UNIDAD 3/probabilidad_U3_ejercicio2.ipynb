```python
# =========================================================
# EJERCICIO 2 - T칩mbola de premios
# =========================================================

import matplotlib.pyplot as plt
import numpy as np

# Valores de la variable aleatoria
x = np.array([0, 10, 20, 50])
p = np.array([0.50, 0.30, 0.15, 0.05])

# 1. Verificar suma
print("Suma de probabilidades:", np.sum(p))

# 2. Valor esperado
E_X = np.sum(x * p)
print("Valor esperado E(X):", E_X)

# 3. Varianza y desviaci칩n est치ndar
E_X2 = np.sum(x**2 * p)
Var_X = E_X2 - E_X**2
Desv_X = np.sqrt(Var_X)
print("Varianza:", Var_X)
print("Desviaci칩n est치ndar:", Desv_X)

# 4. Funci칩n acumulada
F = np.cumsum(p)
print("Funci칩n acumulada F(x):", F)

# 5. Gr치fica de FDA
plt.step(x, F, where='post', color='green', label='F(x)')
plt.title("Funci칩n de Distribuci칩n Acumulada - Ejercicio 2")
plt.xlabel("Premio (pesos)")
plt.ylabel("Probabilidad acumulada")
plt.grid(True)
plt.legend()
plt.show()
```
