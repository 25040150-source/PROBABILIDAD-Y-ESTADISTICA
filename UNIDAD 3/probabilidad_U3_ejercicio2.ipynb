¿que significa el valor esperado en este contexto?

El valor esperado
𝐸(𝑋)
E(X) es 8.5 pesos, lo que significa que en promedio, el jugador ganará 8.5 pesos por cada boleto comprado.
Esto representa la ganancia promedio que se obtendría si el juego se repitiera muchas veces, considerando todas las probabilidades de los distintos premios.

¿el juego es favorable para el jugador?

No, el juego no es favorable para el jugador. Aunque existe la posibilidad de ganar hasta 50 pesos, la mayoría de las veces se gana poco o nada (la probabilidad de no ganar nada es del 50%).
Además, la desviación estándar es alta (11.95), lo que indica una gran variabilidad en los resultados. Esto significa que el jugador tiene más posibilidades de perder que de obtener una ganancia constante. En otras palabras, el juego está diseñado para que, en promedio, el organizador gane más que el participante.


```python
# =========================================================
# EJERCICIO 2 - Tómbola de premios
# =========================================================

import matplotlib.pyplot as plt
import numpy as np

# Valores de la variable aleatoria
x = np.array([0, 10, 20, 50])
p = np.array([0.50, 0.30, 0.15, 0.05])

# 1. Verificar suma
print("Suma de probabilidades:", np.sum(p))

# 2. Valor esperado
E_X = np.sum(x * p)
print("Valor esperado E(X):", E_X)

# 3. Varianza y desviación estándar
E_X2 = np.sum(x**2 * p)
Var_X = E_X2 - E_X**2
Desv_X = np.sqrt(Var_X)
print("Varianza:", Var_X)
print("Desviación estándar:", Desv_X)

# 4. Función acumulada
F = np.cumsum(p)
print("Función acumulada F(x):", F)

# 5. Gráfica de FDA
plt.step(x, F, where='post', color='green', label='F(x)')
plt.title("Función de Distribución Acumulada - Ejercicio 2")
plt.xlabel("Premio (pesos)")
plt.ylabel("Probabilidad acumulada")
plt.grid(True)
plt.legend()
plt.show()
```
