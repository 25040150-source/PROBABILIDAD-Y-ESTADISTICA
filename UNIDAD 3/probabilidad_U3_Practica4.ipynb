```python
import math
import numpy as np
import matplotlib.pyplot as plt
x=[5,6,7,8,9,10]
p=[0.05,0.10,0.20,0.30,0.25,0.10]

F = np.cumsum(p)

for xi, fi in zip(x, F):
    print(f"F({xi}) = {fi:.2f}")

#graficar
plt.step(x,F,where='post')
plt.title('Función de Distribución Acumulada (FDA)')
plt.xlabel('CALIFICACION')
plt.ylabel('PROBABILIDAD ACUMULADA')
plt.grid(True)
plt.ylim(0,1.05)
plt.show()
```

```python
#GRAFICCA DE UNA DISTRIBUCION DE PROBABILIDAD DISCRETA CON MEDIA, VARIANZA Y DESVIACION ESTANDAR
#datos valores de la variable aleatoria y sus probabilidades
x = np.array([5,6,7,8,9,10])
p = np.array ([0.05,0.10,0.20,0.30,0.25,0.10])

#calculo de valor esperado
esperanza = np.sum(x * p)
#calculo de la esperanza del cuadrado
esperanza_cuadrado = np.sum(x**2 * p)
#calculo de la varianza y desviacion estandar
varianza = esperanza_cuadrado - esperanza**2
desviacion = np.sqrt(varianza)

#mostrar reultados en la consola
print(f"E(X) = {esperanza:.2f}")
print(f"E(X ^ 2) = {esperanza_cuadrado:.2f}")
print(f"Var(X) = {varianza:.2f}")
print(f"Std(X) = {desviacion:.2f}")

#crear grafica tipo 'stem' (tallos y hojas)
plt.stem(x,p, basefmt=" ",linefmt= 'b-')
plt.title('Distribución de Probabilidad Discreta')
plt.xlabel('valores de la variable aleatoria X')
plt.ylabel('P(X)')
plt.grid(True, linestyle='--', alpha=0.6)
#dibujar linea vertical en la media
plt.axvline(x=esperanza, color='red', linestyle='--', label=f"Media (E[X]) = {esperanza:.2f}")

#Resaltar el rango de +- desviacion estandar
plt.axvspan(esperanza - desviacion, esperanza + desviacion, color='red', alpha=0.2, label=f"1 desviacion estandar ={desviacion:.2f}")

#mostrar leyenda y limites
plt.legend()
plt.ylim(0,max(p) + 0.05)
plt.show()
```
